Dùng dữ liệu X và y vừa tách để chuẩn bị các tập train/test:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split

# Đọc dữ liệu
column_names = ['ID', 'Diagnosis'] + [f'feature_{i}' for i in range(1, 31)]
data_path = 'F:/Project-AI/breast+cancer+wisconsin+diagnostic/wdbc.data'
data = pd.read_csv(data_path, header=None, names=column_names)

# Tách features và labels
X = data.drop(['ID', 'Diagnosis'], axis=1)
y = data['Diagnosis'].map({'M': 0, 'B': 1})

# Các tỷ lệ train/test
splits = [(0.4, 0.6), (0.6, 0.4), (0.8, 0.2), (0.9, 0.1)]
datasets = {}

# Chia dữ liệu
for train_size, test_size in splits:
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, train_size=train_size, test_size=test_size, 
        stratify=y, shuffle=True, random_state=42
    )
    datasets[f'train_{int(train_size*100)}'] = (X_train, y_train)
    datasets[f'test_{int(test_size*100)}'] = (X_test, y_test)

# Tạo DataFrame để trực quan hóa
plot_data = []
for name, (X_data, y_data) in datasets.items():
    temp_df = pd.DataFrame({'Class': y_data, 'Dataset': name})
    plot_data.append(temp_df)

# Kết hợp tất cả dữ liệu vào một DataFrame
plot_df = pd.concat(plot_data)

# Trực quan hóa
plt.figure(figsize=(12, 8))
sns.countplot(data=plot_df, x='Class', hue='Dataset')
plt.title("Class Distribution Across Train/Test Splits")
plt.xlabel("Class (0: Malignant, 1: Benign)")
plt.ylabel("Count")
plt.show()
```

Xây dựng bộ phân loại decision tree:

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import graphviz

classifiers = {}
feature_names = X.columns  # Tên đặc trưng: feature_1, feature_2,...
class_names = ['Malignant', 'Benign']  # Nhãn lớp

for train_size in [40, 60, 80, 90]:
    X_train, y_train = datasets[f'train_{train_size}']
    clf = DecisionTreeClassifier(criterion='entropy', random_state=42)
    clf.fit(X_train, y_train)
    classifiers[train_size] = clf

    # Trực quan hóa (ví dụ với 80/20)
    if train_size == 80:
        dot_data = export_graphviz(clf, feature_names=feature_names, 
                                   class_names=class_names, filled=True)
        graph = graphviz.Source(dot_data)
        graph.render("breast_cancer_tree_80_20", view=True)
```

Đánh giá các bộ phân loại decision tree:

Phân tích độ chính xác với các giá trị max_depth khác nhau theo tỷ lệ chia 40/60:

```python
from sklearn.metrics import classification_report, confusion_matrix

for train_size in [40, 60, 80, 90]:
    X_test, y_test = datasets[f'test_{100-train_size}']
    y_pred = classifiers[train_size].predict(X_test)
    print(f"Train/Test: {train_size}/{100-train_size}")
    print(classification_report(y_test, y_pred, target_names=class_names))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

Phân tích độ chính xác với các giá trị max_depth khác nhau theo tỷ lệ chia 80/20.

```python
from sklearn.metrics import accuracy_score

X_train, y_train = datasets['train_80']
X_test, y_test = datasets['test_20']
depths = [None, 2, 3, 4, 5, 6, 7]
accuracies = []

for depth in depths:
    clf = DecisionTreeClassifier(criterion='entropy', max_depth=depth, random_state=42)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)
    # Trực quan hóa
    dot_data = export_graphviz(clf, feature_names=feature_names, 
                               class_names=class_names, filled=True)
    graph = graphviz.Source(dot_data)
    graph.render(f"breast_cancer_tree_depth_{depth}", view=True, format='png')

# Bảng và biểu đồ
df = pd.DataFrame({'max_depth': depths, 'Accuracy': accuracies})
print(df)
plt.plot(depths, accuracies, marker='o')
plt.title("Accuracy vs Max Depth (Breast Cancer)")
plt.xlabel("Max Depth")
plt.ylabel("Accuracy")
plt.show()
```
